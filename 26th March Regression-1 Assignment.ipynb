```python
#Regression-1 Assignment
"""Q1. Explain the difference between simple linear regression and multiple linear regression. Provide an
example of each."""

Ans: Simple linear regression is a type of regression analysis that models the relationship between one 
independent variable and one dependent variable. An example of simple linear regression would be 
predicting the price of a house based on its square footage. 

Multiple linear regression is a type of regression analysis that models the relationship between two or 
more independent variables and one dependent variable. An example of multiple linear regression would be 
predicting the price of a house based on its square footage, number of bedrooms, and number of bathrooms.

"""Q2. Discuss the assumptions of linear regression. How can you check whether these assumptions hold in
a given dataset?"""
Ans: The assumptions of linear regression are:
1. Linearity: The relationship between the independent and dependent variables should be linear.
2. Homoscedasticity: The variance of the residuals should be constant across all values of the independent
variable.
3. Normality: The residuals should be normally distributed.
4. Independence: The residuals should be independent of each other.
5. No multicollinearity: There should be no correlation between the independent variables.

To check whether these assumptions hold in a given dataset, you can use graphical methods such as scatter 
plots, histograms, and boxplots. You can also use statistical tests such as the Shapiro-Wilk test for 
normality and the Durbin-Watson test for independence. Additionally, you can use correlation matrices to 
check for multicollinearity.

"""Q3. How do you interpret the slope and intercept in a linear regression model? Provide an example using
a real-world scenario."""
Ans: The slope and intercept in a linear regression model represent the relationship between two variables.
The slope is the rate of change of one variable with respect to the other, and the intercept is the point 
at which the line crosses the y-axis. For example, if we were looking at a linear regression model of the 
relationship between the number of hours studied and the grade on a test, the slope would represent how 
much the grade increases for each additional hour studied, and the intercept would represent the grade 
when no hours are studied.  

"""Q4. Explain the concept of gradient descent. How is it used in machine learning?"""
Ans: Gradient descent is an optimization algorithm used to find the values of parameters (coefficients) of 
a function (f) that minimizes a cost function (cost). It is used in machine learning to update the weights
and bias of a model in order to minimize the cost function. The algorithm works by taking the derivative 
(slope) of the cost function with respect to each parameter, and then updating the parameters in the 
opposite direction of the gradient. This process is repeated until a local minimum is found. The learning 
rate determines how quickly the algorithm converges to a local minimum.

"""Q5. Describe the multiple linear regression model. How does it differ from simple linear regression?"""
Ans: Multiple linear regression is a type of linear regression model that uses two or more independent 
variables to predict a single dependent variable. It is used to analyze the relationship between multiple 
independent variables and a single dependent variable. Unlike simple linear regression, which only uses 
one independent variable to predict the dependent variable, multiple linear regression uses multiple 
independent variables to predict the dependent variable. The model is represented by a linear equation, 
where each independent variable is multiplied by a coefficient and added together to predict the dependent
variable. The coefficients are estimated using the least squares method.

"""Q6. Explain the concept of multicollinearity in multiple linear regression. How can you detect and
address this issue?"""
Ans: 
Multicollinearity is a phenomenon in which two or more predictor variables in a multiple linear regression 
model are highly correlated. This can lead to unreliable and unstable estimates of regression coefficients,
as well as inaccurate hypothesis tests.

Multicollinearity can be detected by looking at the correlation matrix of the predictor variables. If two 
or more variables have a correlation coefficient greater than 0.7, then multicollinearity is likely 
present.

To address multicollinearity, one can remove one of the highly correlated predictor variables, or use 
regularization techniques such as ridge regression or lasso regression. Additionally, one can transform 
the predictor variables to reduce the correlation between them.

"""Q7. Describe the polynomial regression model. How is it different from linear regression?"""
Ans: Polynomial regression is a type of linear regression in which the relationship between the independent 
variable x and the dependent variable y is modeled as an nth degree polynomial. Polynomial regression can
be used to fit non-linear relationships between variables. It is different from linear regression in that
it can fit more complex relationships between variables, such as quadratic, cubic, and higher-order 
polynomials.

Polynomial regression can also be used to fit data that is not linear, such as data with multiple peaks or 
valleys. Linear regression is limited to fitting linear relationships between variables, and cannot be 
used to fit non-linear relationships.

"""Q8. What are the advantages and disadvantages of polynomial regression compared to linear
regression? In what situations would you prefer to use polynomial regression?"""
Ans: Advantages of polynomial regression compared to linear regression: 
1. Polynomial regression can fit more complex data than linear regression. 
2. It can capture non-linear relationships between the independent and dependent variables. 
3. It can provide more accurate predictions than linear regression.

Disadvantages of polynomial regression compared to linear regression: 
1. It can be computationally expensive and time consuming. 
2. It can be prone to overfitting if the degree of the polynomial is too high. 
3. It can be difficult to interpret the results.

You would prefer to use polynomial regression when the data is non-linear and linear regression is not 
able to capture the relationship between the independent and dependent variables.
```
